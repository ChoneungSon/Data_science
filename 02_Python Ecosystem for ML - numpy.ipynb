```python
# vstack((a, b)) : 수직으로 쌓는다.
a = np.array([1, 2, 3,])
b = np.array([2, 3, 4,])
np.vstack((a, b))
```

```python
# hstack((a, b)) : 수평으로 쌓는다.
a = np.array([[1], [2], [3],])
b = np.array([[2], [3], [4],])
np.hstack((a, b))
```

```python
# concatenate((a, b), axis=?) : 축을 기준으로 붙인다.
# 2차원 일 때, row 가 0이고, column 이 1이므로,
# concatenate((a, b), axis=0) == vstack((a, b))
# concatenate((a, b), axis=1) == hstack((a, b))

a.T # trnaspose
```
