```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
test_df = pd.read_csv('Titanic/test.csv')
train_df = pd.read_csv('Titanic/train.csv')
```

```python
# 인덱스 지정
train_df.set_index('PassengerId', inplace=True)
test_df.set_index('PassengerId', inplace=True)
```

```python
test_df.head(1)
```

```python
train_index = train_df.index
test_index = test_df.index
```

```python
y_train_df = train_df.pop('Survived')
```

```python
y_train_df[:10]
```

```python
# 숫자들을 소숫점 3자리 까지만 표시
pd.set_option('display.float_format', lambda x: '%.2f' %x)
```

```python
# 결측치 찾기
train_df.isnull().sum()/len(train_df)*100
```

```python
# Drop Cabin
del train_df['Cabin']
del test_df['Cabin']
```

```python
# 전처리 해주기 위해서는 같은 방식으로 해줘야 하기 때문에 두 데이터를 합친다.
all_df = train_df.append(test_df)
all_df
```

```python
(all_df.isnull().sum()/len(all_df)).plot(kind='bar')
plt.show()
```

```python
len(all_df)
```

```python
del all_df['Name']
```

```python
del all_df['Ticket']
```

```python
all_df
```

```python
all_df['Sex'] = all_df['Sex'].replace({"male":0, 'female':1})
```

```python
all_df
```

```python
# 나오는 내용들을 표시
all_df['Embarked'].unique()
```

```python
all_df['Embarked'] = all_df['Embarked'].replace({'S':0, 'C':1, 'Q':2, np.nan:99})
```

```python
all_df['Embarked'].unique()
```

```python
pd.get_dummies(all_df['Embarked'], prefix='embarked')
```

```python
matrix_df = pd.merge(
    all_df, pd.get_dummies(all_df['Embarked'], prefix='embarked'),
    left_index=True, right_index=True
)
```

```python
# 상관관계 살피기
matrix_df.corr()
```

```python
all_df.groupby('Pclass')["Age"].mean()
```

```python
all_df.groupby('Sex')['Age'].mean()
```

```python
# Pclass 가 1이고, Age가 null인 것의 Age를 불러온다
all_df.loc[(all_df['Pclass'] == 1) & (all_df['Age'].isnull()), 'Age']
```

```python
all_df['Age'].fillna(
    all_df.groupby('Pclass')['Age'].transform('mean'), inplace=True
)
```

```python
all_df
```

```python
all_df.isnull().sum()
```

```python
all_df['Fare'].fillna(
    all_df.groupby('Pclass')['Fare'].transform('mean'), inplace=True
)
```

```python
all_df.isnull().sum()
```

```python
del all_df['Embarked']
```

```python
all_df['Pclass'] = all_df['Pclass'].replace({1:'A', 2:'B', 3:'C'})
```

```python
all_df = pd.get_dummies(all_df)
```

```python
all_df = pd.merge(
    all_df, matrix_df[['embarked_0', 'embarked_1', 'embarked_2', 'embarked_99']],
    left_index=True, right_index=True,
)
```

```python
train_df = all_df[all_df.index.isin(train_index)]
test_df = all_df[all_df.index.isin(test_index)]
```

# Build Model

```python
x_data = np.array(train_df)
y_data = np.array(y_train_df)
```

```python
x_data.shape, y_data.shape
```

```python
y_data
```

```python
from sklearn.linear_model import LogisticRegression
cls = LogisticRegression()
cls.fit(x_data, y_data)
```

```python
cls.intercept_
```

```python
cls.coef_
```

```python
cls.predict(test_df.values)
```

```python
test_df.index
```

```python
x_test = np.array(test_df)
y_test = cls.predict(x_test)
y_test
```

```python
result = np.concatenate(
    (test_index.values.reshape(-1,1), cls.predict(x_test).reshape(-1,1)),
    axis = 1
)
```

```python
df_submssion = pd.DataFrame(result, columns=['PassengerId', 'Survived'])
```

```python
df_submssion
```

```python
df_submssion.to_csv("Titanic/submission_result.csv", index=False)
```
