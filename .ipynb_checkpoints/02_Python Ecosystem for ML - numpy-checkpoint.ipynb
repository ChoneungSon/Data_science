# Feature
- 머신러닝에서 데이터의 특징을 나타내는 변수
- feature, 독립변수, input 변수 등은 동일 의미로 사용됨
- 일반적으로 Table 상에 Data를 표현할 때, Column을 의미
- 하나의 data instance (실제 데이터)는 feature vector로 표현



```python
# numpy
# 파이썬은 굉장히 느리기 때문에 패키지를 이용하여 계산하여야 한다.

import numpy as np
```

```python
test_arr = np.array([1,4, 5, 8], float)
print(test_arr)
type(test_arr[3])
# numpy.float64는 64bit 숫자를 의미함
```

```python
test_arr.dtype
```

```python
# 크기가 4인 1차원 텐서
test_arr.shape
```

```python
# 차원이 하나씩 늘어날 때마다, 앞에 쌓인다.
test_arr.ndim # 차원
test_arr.size # data의 개수
```

```python
# 메모리 관련 이슈는 나중에 난이도가 올라가면
```

```python
# 1차원으로 만든 뒤 순서대로 reshape
test_matrix = [[1,2,3,4,],[5,6,7,8,]]
np.array(test_matrix).reshape(2,2,2)
```

```python
# row의 크기는 모르지만 column의 크기를 정하고 싶을 때
np.array(test_matrix).reshape(-1, 2)
```

```python
np.array(test_matrix).shape
```

```python
# 1차원으로 펴줄 때 사용, reshape 써도 된다.
np.array(test_matrix).flatten()
```

```python
test_ex = np.array(test_matrix, int)
# matlab과 동일하게 이용가능
test_ex[0, 0]
```

```python
# 슬라이싱도 matlab과 동일하게 가능
test_ex[:,1]
```

```python
# range 와 동일한 작용
np.arange(30).reshape(6, 5)
```
