```python
GENRE_WEIGHT = 0.1

def pearsonR(s1, s2):
    s1_c = (s1 - s1.mean())/s1.mean()
    s2_c = (s2 - s2.mean())/s2.mean()
    return np.sum(s1_c * s2_c) / np.sqrt(np.sum(s1_c ** 2) * np.sum(s2_c ** 2))

def find_genre(id):
    genre_query = cur.execute("SELECT DISTINCT genre_id FROM movies_movie_genres WHERE movie_id = (%d)", (id))
    cols = [column[0] for column in genre_query.description]
    return pd.DataFrame.from_records(data=genre_query.fetchall(), columns=cols)

def calcul_same_genre(id1, id2):
    return pd.Series(list(set(find_genre(id1).genre_id) & set(find_genre(id2).genre_id))).count()
    

def recommend(Id, matrix, n):
    
    result = []
    
    for side_id in matrix.columns:
        
        if side_id == Id:
            continue

        # rating comparison
        cor = pearsonR(matrix[Id], matrix[side_id])
        
        if np.isnan(cor):
            result.append((side_id, 0))
        else:
            result.append((side_id, cor))
            
    result.sort(key=lambda r: -r[1])
    print(result)
    result = result[0][0]
    similar_query = cur.execute("SELECT id FROM movies_movie WHERE NOT id IN (SELECT DISTINCT movie_id FROM movies_review WHERE user_id = {}) ORDER BY -popularity".format(result))
    cols = [column[0] for column in query.description]
    similar = pd.DataFrame.from_records(data=query.fetchall(), columns=cols)
    
    return similar[:n]
```
